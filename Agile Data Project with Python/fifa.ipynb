```python

# Filtrar as linhas onde o jogador contém "Arantes"
pele_gols = Jogadores[Jogadores['Player Name'].str.contains('Arantes', case=False, na=False)].copy()

# Adicionar uma coluna para contar os gols
pele_gols['Gol Count'] = pele_gols['Event'].str.count('G')

# Filtrar os jogos onde Pelé fez gols
jogos_pele = jogos[jogos['MatchID'].isin(pele_gols['MatchID'])]

# Mesclar os DataFrames para obter informações adicionais
pele_gols_info = pd.merge(pele_gols, jogos_pele[['MatchID', 'Stadium', 'City']], left_on='MatchID', right_on='MatchID', how='left')

# Criar um DataFrame para contar cada ocorrência de 'G' individualmente
gol_count_df = pele_gols_info['Event'].str.extractall('(G\d+)').reset_index()
gol_count_df.columns = ['match_index', 'event_index', 'Gol Event']

# Se existirem linhas sem correspondência, adicione valores padrão para 'Stadium' e 'City'
gol_count_df[['Stadium', 'City']] = pele_gols_info[['Stadium', 'City']].fillna('Desconhecido')

# Agrupar e contar os gols por estádio e cidade
gol_count_by_stadium_city = gol_count_df.groupby(['Stadium', 'City']).size().reset_index(name='Gol Count')

# Imprimir os estádios, cidades e contagem de gols onde Pelé fez gols sem exibir o índice
print("Estádios, cidades e contagem de gols onde Pelé fez gols:")
print(gol_count_by_stadium_city.to_string(index=False))
```

```python
import pandas as pd
from tabulate import tabulate

# Leitura dos DataFrames Jogos, Jogadores e Tabela
jogos = pd.read_csv('WorldCupMatches.csv', encoding='latin1')
jogadores = pd.read_csv('WorldCupPlayers.csv', encoding='latin1')
tabela = pd.read_csv('WorldCups.csv', encoding='latin1')

# Filtrar as linhas onde o jogador contém "Arantes"
pele_gols = jogadores[jogadores['Player Name'].str.contains('Arantes', case=False, na=False)].copy()

# Adicionar uma coluna para contar os gols
pele_gols['Gol Count'] = pele_gols['Event'].str.count('G')

# Filtrar os jogos onde Pelé fez gols
jogos_pele = jogos[jogos['MatchID'].isin(pele_gols['MatchID'])]

# Mesclar os DataFrames para obter informações adicionais
pele_gols_info = pd.merge(pele_gols, jogos_pele[['MatchID', 'Stadium', 'City', 'Attendance']], left_on='MatchID', right_on='MatchID', how='left')

# Criar um DataFrame para contar cada ocorrência de 'G' individualmente
gol_count_df = pele_gols_info['Event'].str.extractall('(G\d+)').reset_index()
gol_count_df.columns = ['match_index', 'event_index', 'Gol Event']

# Se existirem linhas sem correspondência, adicione valores padrão para 'Stadium', 'City', e 'Attendance'
gol_count_df[['Stadium', 'City', 'Attendance']] = pele_gols_info[['Stadium', 'City', 'Attendance']].fillna('Desconhecido')

# Agrupar e contar os gols por estádio, cidade e somar a assistência
gol_count_by_stadium_city = gol_count_df.groupby(['Stadium', 'City'])[['Gol Event', 'Attendance']].agg({'Gol Event': 'count', 'Attendance': 'sum'}).reset_index()

# Criar um DataFrame organizado
resultado_df = pd.DataFrame({
    'Estádio': gol_count_by_stadium_city['Stadium'],
    'Cidade': gol_count_by_stadium_city['City'],
    'Contagem de Gols': gol_count_by_stadium_city['Gol Event'],
    'Total de Pessoas': gol_count_by_stadium_city['Attendance']
})
# Adicionar linhas para a contagem total de gols e o total de pessoas
total_gols = resultado_df['Contagem de Gols'].sum()
total_pessoas = resultado_df['Total de Pessoas'].sum()

resultado_df.loc[len(resultado_df)] = ['Total', '', total_gols, total_pessoas]

# Imprimir a tabela organizada sem os totais
print("Estádios, cidades, contagem de gols e total de pessoas que viram os gols de Pelé:")
print(tabulate(resultado_df.iloc[:-1, :], headers='keys', tablefmt='psql', showindex=False))

# Adicionar linha horizontal antes da última linha com totais
print("\n" + "-"*50)
print(tabulate(resultado_df.tail(1), headers='keys', tablefmt='psql', showindex=False))
```
