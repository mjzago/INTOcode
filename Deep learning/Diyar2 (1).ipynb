```python


# Load the MNIST dataset
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()

# Preprocess the data
train_images = train_images.reshape((60000, 28, 28, 1)).astype('float32') / 255
test_images = test_images.reshape((10000, 28, 28, 1)).astype('float32') / 255

train_labels = to_categorical(train_labels)
test_labels = to_categorical(test_labels)

# Build the MNIST model
mnist_model = Sequential()
mnist_model.add(Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)))
mnist_model.add(MaxPooling2D((2, 2)))
mnist_model.add(Conv2D(64, (3, 3), activation='relu'))
mnist_model.add(MaxPooling2D((2, 2)))
mnist_model.add(Conv2D(128, (3, 3), activation='relu'))
mnist_model.add(MaxPooling2D((2, 2)))
mnist_model.add(Flatten())
mnist_model.add(Dense(512, activation='relu'))
mnist_model.add(Dense(128, activation='relu'))
mnist_model.add(Dense(10, activation='softmax'))

# Compile the model
mnist_model.compile(optimizer='adam',
                    loss='categorical_crossentropy',
                    metrics=['accuracy'])

# Train the model
history_mnist = mnist_model.fit(train_images, train_labels, epochs=10, batch_size=8, validation_data=(test_images, test_labels))

# Plot the training history for the MNIST model
plt.figure(figsize=(12, 4))

# Plot training & validation loss values
plt.subplot(1, 2, 1)
plt.plot(history_mnist.history['loss'])
plt.plot(history_mnist.history['val_loss'])
plt.title('MNIST Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend(['Train', 'Test'], loc='upper right')

# Plot training & validation accuracy values
plt.subplot(1, 2, 2)
plt.plot(history_mnist.history['accuracy'])
plt.plot(history_mnist.history['val_accuracy'])
plt.title('MNIST Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend(['Train', 'Test'], loc='lower right')

plt.tight_layout()
plt.show()


```

```python
#MODEL 3

# Load the CIFAR-10 dataset
(train_images, train_labels), (test_images, test_labels) = cifar10.load_data()

# Preprocess the data
train_images = train_images.astype('float32') / 255
test_images = test_images.astype('float32') / 255

train_labels = to_categorical(train_labels, 10)
test_labels = to_categorical(test_labels, 10)

# Build a more complex CIFAR-10 model (Model 3)
model3_cifar10_model = Sequential()
model3_cifar10_model.add(Conv2D(64, (3, 3), activation='relu', input_shape=(32, 32, 3)))
model3_cifar10_model.add(Conv2D(64, (3, 3), activation='relu'))
model3_cifar10_model.add(MaxPooling2D((2, 2)))
model3_cifar10_model.add(Dropout(0.25))

model3_cifar10_model.add(Conv2D(128, (3, 3), activation='relu'))
model3_cifar10_model.add(Conv2D(128, (3, 3), activation='relu'))
model3_cifar10_model.add(MaxPooling2D((2, 2)))
model3_cifar10_model.add(Dropout(0.25))

model3_cifar10_model.add(Flatten())
model3_cifar10_model.add(Dense(512, activation='relu'))
model3_cifar10_model.add(Dropout(0.5))
model3_cifar10_model.add(Dense(256, activation='relu'))
model3_cifar10_model.add(Dense(10, activation='softmax'))

# Compile the model
model3_cifar10_model.compile(optimizer='adam',
                             loss='categorical_crossentropy',
                             metrics=['accuracy'])

# Train the model
history_model3_cifar10 = model3_cifar10_model.fit(train_images, train_labels, epochs=10, batch_size=64, validation_data=(test_images, test_labels))

# Plot the training history for Model 3
plt.figure(figsize=(12, 4))

# Plot training & validation loss values
plt.subplot(1, 2, 1)
plt.plot(history_model3_cifar10.history['loss'])
plt.plot(history_model3_cifar10.history['val_loss'])
plt.title('Model 3 CIFAR-10 Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend(['Train', 'Test'], loc='upper right')

# Plot training & validation accuracy values
plt.subplot(1, 2, 2)
plt.plot(history_model3_cifar10.history['accuracy'])
plt.plot(history_model3_cifar10.history['val_accuracy'])
plt.title('Model 3 CIFAR-10 Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend(['Train', 'Test'], loc='lower right')

plt.tight_layout()
plt.show()
```

```python
# SUMARY

import matplotlib.pyplot as plt

def plot_training_history(histories, names, title='Model Training Comparison'):
    plt.figure(figsize=(12, 4))

    # Plot training loss values
    plt.subplot(1, 2, 1)
    for history in histories:
        plt.plot(history.history['loss'])
    plt.title('Training Loss Comparison')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend(names, loc='upper right')

    # Plot training accuracy values
    plt.subplot(1, 2, 2)
    for history in histories:
        plt.plot(history.history['accuracy'])
    plt.title('Training Accuracy Comparison')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend(names, loc='lower right')

    plt.tight_layout()
    plt.suptitle(title, y=1.02)
    plt.show()

# Call the function with the model histories and names
plot_training_history([history_cifar10, history_model2, history_model3_cifar10, history_model4_cifar10, history_model5_cifar10],
                      ['Model 1', 'Model 2', 'Model 3', 'Model 4', 'Model 5'])
```
