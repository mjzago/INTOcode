```python
import codecs
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords
from nltk.probability import FreqDist
import glob

# Configuration for NLTK (download the necessary resources)
nltk.download('punkt')
nltk.download('stopwords')

# Folder where the text files are located
folder = 'Firmen/'

# List of stopwords in English (or in the appropriate language)
stop_words = set(stopwords.words('english'))

# Initialize a counter for nouns
nouns_counter = FreqDist()

# List to store all sentences from all texts
all_sentences = []

# Using the glob module to get the list of files in the folder
files = glob.glob(folder + '*.txt')

for file in files:
    with codecs.open(file, 'r', 'utf8') as f:
        text = f.read()
        # Remove '=' characters and whitespace at the beginning and end of each line
        lines = [line.strip('= \n ') for line in text.split('\n') if line.strip('= \n ')]
        text = ' '.join(lines)

        # Tokenize into sentences
        sentences = sent_tokenize(text)

        # Append sentences to the list
        all_sentences.extend(sentences)

        for sentence in sentences:
            # Tokenize into words and remove stopwords
            words = word_tokenize(sentence)
            words = [word.lower() for word in words if word.isalpha() and word.lower() not in stop_words]

            # POS tagging to identify nouns (NN, NNS)
            tagged_words = nltk.pos_tag(words)
            nouns = [word for word, tag in tagged_words if tag.startswith('NN')]

            # Update the noun counter
            nouns_counter.update(nouns)

# Display the 10 most common nouns across all texts
print("Most important nouns in all texts:")
for word, freq in nouns_counter.most_common(10):
    print(f"{word}: {freq} times")

# Display the list of all sentences from all texts
print("\nAll Sentences:")
for i, sentence in enumerate(all_sentences, 1):
    print(f"{i}: {sentence}")
```
